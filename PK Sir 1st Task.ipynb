```python
merged = geo.set_index('st_nm').merge(data, on = 'st_nm', how = 'left')
merged.head()
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis('off')
ax.set_title('Students Admission', fontdict={'fontsize': '25', 'fontweight' : '10'})
merged.plot(column='Count', cmap='YlGnBu', linewidth=0.8, ax=ax, edgecolor='0', legend=True,markersize=[39.739192, -104.990337])
```

```python
variable = 'Count'
vmin, vmax = 120, 220
fig, ax = plt.subplots(1, figsize=(10, 10))
merged.plot(column=variable, cmap='YlOrRd')
ax.axis('off')
```

```python
sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm)
fig, ax = plt.subplots(1, figsize=(10, 6))
merged.plot(column=variable, cmap='Greens',linewidth=0.8, ax=ax, edgecolor='0.8')
print(cbar)
fig.savefig('testmap.png', dpi=300)
```
